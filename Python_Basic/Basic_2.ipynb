###### 2) zip
- 2개의 리스트를 key, value 형태로 묶어주는 함수

````python
# 2개의 리스트를 매핑시켜 딕셔너리 형태로 변경하세요
subjects = ["korean", "english", "math", "science"]
points = [100, 80, 90, 60]

dic = {}
for idx in range(len(subjects)):
    dic[subjects[idx]] = points[idx]

print(dic)
````

````python
dic ={}
for subject, point in zip(subjects, points):
    dic[subject] = point

print(dic)
````

````python
dict(zip(subjects, points))
````

````python
tuple(zip(subjects, points))
````

- 리스트 형태의 데이터를 index와 value 값을 동시에 사용할수 있는 함수

###### 1) 5명의 이름을 입력 받아 kim인 성을 가진 사람의 수를 출력
- [성(공백)이름] 형태로만 입력(ex : park doojin)
- names = ['kim sung', 'Kim daegu', 'park sangjin', 'lee minsung', 'jung sunin']

````python
names = ['kim sung', 'Kim daegu', 'park sangjin', 'lee minsung', 'jung sunin']
count = 0
name_list =[]
for name in names:
    last_name = name.split(" ")[0].lower()
    if last_name == "kim":
        count += 1
        name_list.append(name)
        
        
print(name_list, count)
````

##### 3) up down으로 숫자 맞추기 게임
- 랜덤으로 숫자(1~100)를 발생시킴
- 숫자를 입력받아 랜덤숫자가 입력받은 숫자보다 크면 Up을 출력, 작으면 Down을 출력
- 랜덤 숫자를 맞추면 몇번만에 맞췄는지 카운트를 알려줌
- 랜덤수 발생 방법
```
import random
random_number = random.randint(1,100)
```

````python
import random

random_number = random.randint(1,100)
count =0

while True:
    input_number = int(input("please enter the number : "))
    count += 1
    
    if input_number > random_number:
        print("Go Down")
        continue
    elif input_number < random_number:
        print("Go Up")
        continue
    else:
        print("Congratuation!!")
        break
        
print("{}번 만에 맞추었습니다!".format(count))
    
````

###### 5) 로또번호 출력
- 1~45의 6자리의 로또번호를 출력

````python
import random

result =[]
while len(result) < 6:
    
    # 1부터 45까지의 랜덤수발생
    number = random.randint(1,45)
    
    if number not in result:
        result.append(number)

result.sort()
print(result)
````

###### 6) FizzBuzz
- 1에서 1씩 숫자를 입력받은 숫자까지 증가시키며 3의 배수에서는 Fizz 5의 배수에서는 Buzz 가 출력되도록 프로그램을 작성
- 3과 5의 배수에서는 FizzBuzz가 출력

````python
number = int(input("Please Enter the number"))

for i in range(1, number + 1):
    if (i % 3 ==0) & (i % 5 == 0):
        print(i, " FizzBuzz")
    elif i % 3 == 0:
        print(i, " Fizz")
    elif i % 5 == 0:
         print(i, " Buzz")
    else:
        print(i)
    
        
        
````
