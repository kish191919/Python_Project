###### 10) du

- 디렉토리 이름을 지정하지 않으면 현재 디렉토리를 기준으로 디스크 사용량을 출력

```python
# -h : 사람이 읽을 수 있는 숫자로 변환
! du -h 
```

```python
# -s : 요약된 정보 표시 (사용량 합계)
# 서브 디렉토리의 정보는 출력하지 않음
! du -sh /Users/s107166/Desktop/
```

###### 11) ps
- Process Status
- 현재 실행중인 프로세스 목록
- grep명령어와 함께 사용하여 특정 프로세스를 확인하는데 많이 사용
- 옵션
    - e : 모든 프로세스를 보여줌
    - f : full format 으로 보여줌
        - uid(user ID), pid(process ID), ppid(parent ID), TTY(프로세스와 연결된 터미널) 등을 표시
    - l : 풀 포맷정보 외에 F(프로세스 플래그), S(프로세스 상태), PRI(우선순위) 등 더 많은 정보를 보여줌


```python
! ps
```

```python
# -e : 모든 프로세스를 보여줌
! ps -e
```

```python
! ps -e | grep "jupyter"
```

```python
# -f : full format 으로 보여줌
# uid(user ID), pid(process ID), ppid(parent ID), TTY(프로세스와 연결된 터미널) 등을 표시
! ps -f
```

```python
# F(프로세스 플래그), S(프로세스 상태), PRI(우선순위) 등 더 많은 정보보여줌
! ps -l
```
