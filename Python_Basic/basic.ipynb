## 13. Global 함수

````python
# 함수안에서 함수 바깥에 있는 변수의 값을 꼭 바꿔야만 한다면 다음과 같이 함수 이름 앞에 global 키워드를 선언해 주면 된다.
z= 3
def f(x):
    z = 99
    y = x * z
    print("y is ", y)
    print("z is ", z)
    return y

f(10)
print("z is ",z)
````

````python
z= 3
def f(x): 
    global z 
    z = 99
    y = x * z
    print("y is ", y)
    print("z is ", z)
    return y

f(10)
print("z is ",z)
````
