## 6. % 포멧 

````python
# '2 곱하기 3은 6이다.' % 사용
print("%d 곱하기 %d은 %d이다." % (2,3,6))
````

````python
# '철수의 수학 과목 점수는 100점이다.' % 사용
print("%s의 %s 과목 점수는 %d점이다." % ("철수", "수학", 100))
````

````python
# [와 *] 사이에 20칸의 공백이 있다.
print("[%20s]"% "*")
````

````python
# 20칸의 공백의 앞쪽에 A를 인쇄한다.
print("[%-20s]" % "A")
````

````python
# 20칸의 공백의 뒷쪽에 123을 인쇄한다.
print("[%20d]" % 123)
````

````python
# 소숫점 아래 5자리까지만 인쇄한다.
x = 1 / 3.0
print("[%.5f]" % x)
````

````python
# 20칸의 공백의 앞쪽에 소숫점 아래 6자리 출력
print("[%-20.6f]" % x)
````

## 9. Global 함수

````python
# 함수안에서 함수 바깥에 있는 변수의 값을 꼭 바꿔야만 한다면 다음과 같이 함수 이름 앞에 global 키워드를 선언해 주면 된다.
z= 3
def f(x):
    z = 99
    y = x * z
    print("y is ", y)
    print("z is ", z)
    return y

f(10)
print("z is ",z)
````

````python
z= 3
def f(x): 
    global z 
    z = 99
    y = x * z
    print("y is ", y)
    print("z is ", z)
    return y

f(10)
print("z is ",z)
````

## 10. For 구문

````python
# 숫자를 입력 받아 아래와 같이 별찍기 1
star = int(input("Enter number: "))

for i in range(1,star+1):
    print("*"*i)
````

````python
# 숫자를 입력 받아 아래와 같이 별찍기 2
star = int(input("Enter number: "))
for i in range(star,0,-1):
    print("*" * i)
````

````python
# 숫자를 입력 받아 아래와 같이 별찍기 3
star = int(input("Enter number: "))

space = star - 1
for i in range(1, star+1):
    print(" "*space,"*" * i)
    space = space - 1
````

````python
for i in range(1,11):
    print("*"*i)
    if i == 10:
        for i in range(10,0,-1):
            print("*" * i)
````

````python
for i in range(4):
    for j in range(4):
        print(i + j, end=" ")
    print("")
````
